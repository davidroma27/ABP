# Introducción de instancias

Para introducir instancias emplearemos la clase `DataFrame`.

Dentro de la clase se introducen las columnas con los valores que queramos introducir, cada columna separada por comas.
A la hora de añadir el dataframe usamos el método `append()`.

Para que Pandas indique correctamente el índice correspondiente a la nueva tupla debemos añadir como segundo parametro al metodo append `ignore_index=True`.

Cabe destacar que si se introduce una columna nueva que no existe en el dataset, se creará automaticamente cuando se añada la nueva tupla.

```
#Introducimos los valores de cada columna y almacenamos el dataframe en una variable
articulo = pd.DataFrame({'id':['AVpf3txeLJeJML43FN88'],'brand':['Trunktek'],'categories':['Trunk'],'colors':['multicolor'],'dimension':['20 in x 4 in'],'weight':['10 lbs'],'name':['Trunkator'],'recommended':['True'],'rating':[4.0],'reviews.text':['nice product'],'reviews.title':['Trunk']})

#Añadimos el nuevo dataframe al dataset con el metodo append()
electronics = electronics.append(articulo, ignore_index=True)

electronics.tail() #Visualizamos el final del dataset para comprobar que se ha introducido correctamente
```

# Borrar una instancia

Para borrar una instancia ya existente en el dataset emplearemos el metodo `drop()` pasandole como parámetro el nombre o número de tupla que queremos eliminar.

Al ejecutar la eliminación, el método drop muestra el dataset después de borrar la tupla indicada.

```
electronics.drop(7299)
```

# Buscar instancias por atributo

Para buscar una instancia que existe en el dataset emplearemos los metodos `loc` y `iloc`.

`loc`: Usaremos loc para buscar instacias cuando los atributos tienen nombre.

`iloc`: Usaremos iloc para buscar instancias cuando los atributos son índices numéricos.

Si lo que buscamos es filtrar por columnas, es suficiente con pasar el nombre o índice de la columna y el valor que queremos buscar.

Si queremos buscar un valor concreto, les pasaremos los nombres o índices de las filas y columnas correspondientes al valor que queremos buscar.

```
#Filtrado por columnas
electronics.loc[electronics['brand']=='Microsoft']
```

```
#Busqueda de un valor concreto
electronics.loc[4, 'colors']
```

# Modificar una instancia <a id="section4"/>

Si queremos modificar valores de una instancia emplearemos el método `replace()`. Dentro del metodo introduciremos entre corchetes {} los valores que queremos modificar seguidos de ":" y el valor por el que queremos modificar las instancias.

```
electronics.replace({"Microsoft":"Apple"})
```
